```python
import pandas as pd
import duckdb
import os

path = "../gbif_url_dataset.parquet"
if os.path.exists(path):
    df = pd.read_parquet(path)
    display(df)
else:
    display(f"File {path} does not exist")
```

```python
# Get the list of unique dataset names
unique_datasets = df['datasetName'].unique()
print(f"Unique dataset names: {unique_datasets}")
print(f"Total unique datasets: {len(unique_datasets)}")

# Comprehensive analysis
print(f"Total records: {len(df)}")
print(f"Unique dataset names: {df['datasetName'].nunique()}")
print("\nDataset distribution:")
dataset_counts = df['datasetName'].value_counts()
display(dataset_counts)

# Show percentages
print("\nDataset distribution (percentages):")
dataset_percentages = df['datasetName'].value_counts(normalize=True) * 100
display(dataset_percentages)
```
